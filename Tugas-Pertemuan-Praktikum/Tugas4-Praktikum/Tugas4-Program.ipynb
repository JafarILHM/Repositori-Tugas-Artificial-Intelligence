```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca dataset employee.csv
file_path = "employee.csv"  # Pastikan file ini ada di lokasi yang sesuai
df = pd.read_csv(file_path)
```

```python
# Fungsi keanggotaan Trapesium
def trapezoidal_mf(x, a, b, c, d):
    return np.maximum(0, np.minimum((x - a) / (b - a), 1, (d - x) / (d - c)))
```

```python
# Fungsi keanggotaan Gaussian
def gaussian_mf(x, mean, sigma):
    return np.exp(-0.5 * ((x - mean) / sigma) ** 2)
```

```python
# Menentukan rentang nilai untuk Age dan Salary
age_values = np.linspace(df["Age"].min(), df["Age"].max(), 100)
salary_values = np.linspace(df["Salary"].min(), df["Salary"].max(), 100)
```

```python
# Menghitung fungsi keanggotaan untuk Age
age_trap = trapezoidal_mf(age_values, 25, 30, 40, 45)
age_gauss = gaussian_mf(age_values, np.mean(df["Age"]), np.std(df["Age"]))
```

```python
# Plot fungsi keanggotaan Trapesium dan Gaussian untuk Age dan Salary
plt.figure(figsize=(12, 5))

# Fungsi Grafik Keanggotaan untuk Age
plt.subplot(1, 2, 1)
plt.plot(age_values, age_trap, label="Trapesium", color='blue')
plt.plot(age_values, age_gauss, label="Gaussian", color='red', linestyle="dashed")
plt.title("Fungsi Keanggotaan untuk Age")
plt.xlabel("Age")
plt.ylabel("Degree of Membership")
plt.legend()

# Fungsi Grafik Keanggotaan untuk Salary
plt.subplot(1, 2, 2)
plt.plot(salary_values, salary_trap, label="Trapesium", color='blue')
plt.plot(salary_values, salary_gauss, label="Gaussian", color='red', linestyle="dashed")
plt.title("Fungsi Keanggotaan untuk Salary")
plt.xlabel("Salary")
plt.ylabel("Degree of Membership")
plt.legend()

plt.tight_layout()
plt.show()
```
